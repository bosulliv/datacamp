```r
Mode <- function(x) {
  ux <- unique(x)
  ux[which.max(tabulate(match(x, ux)))]
}
```

```r
r1 <- rnorm(10000, 10, sd = 4)
r2 <- runif(10000, -5, 5)
r3 <- c(r1, r2)

d3 = data.frame(r3)
colnames(d3) <- c("observation")
str(d3)

library(ggplot2)
#install.packages("ggthemes")
library(ggthemes)

ggplot(d3, aes(observation)) +
    theme_hc() +
    scale_color_hc() +
    geom_histogram(binwidth = 0.5)
```
